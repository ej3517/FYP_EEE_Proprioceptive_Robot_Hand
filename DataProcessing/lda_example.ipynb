```python
import json
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_wine
np.set_printoptions(precision=4)
sns.set()
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = pd.Categorical.from_codes(wine.target, wine.target_names)

X.head()
```

# Compute the within class and between class scatter matrices

### Within class scatter matrix where c is the total number of different classes
$S_W = \sum_{i=1}^c S_i$

$S_i = \sum_{x\in D_i}^n (x-m_i)(x-m_i)^T$

### where n is the number of sample and x a sample
$m_i = \frac{1}{n_i} \sum_{x\in D_i}^n x_k$

```python
df = X.join(pd.Series(y, name='class'))

class_feature_means = pd.DataFrame(columns=wine.target_names)
for c, rows in df.groupby('class'):
    class_feature_means[c] = rows.mean()
class_feature_means
```

```python
# Then, we plug the mean vectors (mi) into the equation from before 
# in order to obtain the within class scatter matrix.
n_class = class_feature_means.shape[0]
within_class_scatter_matrix = np.zeros((n_class,n_class))

for c, rows in df.groupby('class'):
    rows = rows.drop(['class'], axis=1)
    s = np.zeros((n_class, n_class))
    for index, row in rows.iterrows():
        x, mc = row.values.reshape(n_class,1),class_feature_means[c].values.reshape(n_class,1)
        s += (x - mc).dot((x - mc).T)
    within_class_scatter_matrix += s
```

# Between Class Scatter Matrix

$S_B = \sum_{i=1}^c N_i(x-m_i)(x-m_i)^T$

$m_i = \frac{1}{n_i} \sum_{x\in D_i}^n x_k$

$m = \frac{1}{n} \sum_{i}^n x_i$

```python
feature_means = df.mean()
between_class_scatter_matrix = np.zeros((n_class,n_class))
for c in class_feature_means:
    n = len(df.loc[df['class'] == c].index)
    mc, m = class_feature_means[c].values.reshape(n_class,1), feature_means.values.reshape(n_class,1)
    between_class_scatter_matrix += n * (mc - m).dot((mc - m).T)
```

# Compute the eigenvectors and corresponding eigenvalues for the scatter matrices

$S_W^{-1}S_B$

```python
eigen_values, eigen_vectors = np.linalg.eig(np.linalg.inv(within_class_scatter_matrix).dot(between_class_scatter_matrix))

pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]
pairs = sorted(pairs, key=lambda x: x[0], reverse=True)
for pair in pairs:
    print(pair[0])

eigen_value_sums = sum(eigen_values)
print("\n")
print('Explained Variance')
print("\n")
for i, pair in enumerate(pairs):
    print('Eigenvector {}: {}'.format(i, (pair[0]/eigen_value_sums).real))
    
w_matrix = np.hstack((pairs[0][1].reshape(13,1), pairs[1][1].reshape(13,1))).real

X_lda = np.array(X.dot(w_matrix))

le = LabelEncoder()
y = le.fit_transform(df['class'])

plt.xlabel('LD1')
plt.ylabel('LD2')
plt.scatter(
    X_lda[:,0],
    X_lda[:,1],
    c=y,
    cmap='rainbow',
    alpha=0.7,
    edgecolors='b'
)
```
